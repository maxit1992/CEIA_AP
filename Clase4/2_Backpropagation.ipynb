b- Numerical Differentiation: para el grafo de c√≥mputo que se muestra en la

--------------------------------------------------------------------------------------------------------------------------------
## Punto a

Primero calculamos Y (paso Forward)
$$
a = 1  \\
b = 2 \\
c = 3 \\
d = a + b = 3 \\
e = b - c = -1 \\
Y = d * e = -3 
$$

Ahora aplicamos backpropagation para calcular las derivadas de Y respecto de las entradas

$$
\frac{dY}{da}= \frac{dY}{dd} * \frac{dd}{da} = e * 1 = e = -1
$$
<br>
$$
\frac{dY}{db} = \frac{dY}{dd} * \frac{dd}{db} + \frac{dY}{de} * \frac{de}{db} = e * 1 + d * 1 = e + d = 2
$$
<br>
$$
\frac{dY}{dc} = \frac{dY}{de} * \frac{de}{dc} = d * -1 = -d = -3
$$

## Punto b

$$
Y = d * e = (a + b) * (b - c) = a * b - a * c + b * b - b * c 
$$
<br>
### dY/da
$$
\lim_{h \rightarrow 0} \frac{Y(a+h,b,c)-Y(a,b,c)}{h} = \lim_{h \rightarrow 0} \frac{ [(a + h) * b - (a + h) * c + b * b - b * c ] -  [a * b - a * c + b * b - b * c]}{h} \\
\lim_{h \rightarrow 0} \frac{h * (b - c)}{h} = \lim_{h \rightarrow 0} (b-c) = b - c = -1 
$$
<br>
### dY/db
$$
\lim_{h \rightarrow 0} \frac{Y(a,b+h,c)-Y(a,b,c)}{h} = \lim_{h \rightarrow 0} \frac{ [a * (b + h) - a * c + (b + h) * (b + h) - (b + h) * c ] -  [a * b - a * c + b * b - b * c]}{h} \\
\lim_{h \rightarrow 0} \frac{ h * (a +  2 * b + h + c)}{h} = \lim_{h \rightarrow 0} (a +  2 * b + h + c) = a + 2 * b + c = 2 
$$
<br>
### dY/dc
$$
\lim_{h \rightarrow 0} \frac{Y(a,b,c+h)-Y(a,b,c)}{h} = \lim_{h \rightarrow 0} \frac{ [a * b - a * (c + h) + b * b - b * (c + h)] -  [a * b - a * c + b * b - b * c]}{h} \\
\lim_{h \rightarrow 0} \frac{ h * (a - b)}{h} = \lim_{h \rightarrow 0} (-a - b) =  -a - b = -3
$$
